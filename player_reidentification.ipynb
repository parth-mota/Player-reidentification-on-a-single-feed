```python
import cv2
import numpy as np
from ultralytics import YOLO
import matplotlib.pyplot as plt
from IPython.display import Video, display, clear_output
import os
from pathlib import Path
import time
```

```python
# Load your downloaded YOLOv11 model
model_path = "C:/Users/LENOVO/Downloads/best.pt" 
model = YOLO(model_path)

# Set input and output paths
input_video = "C:/Users/LENOVO/Downloads/15sec_input_720p.mp4" 
output_dir = "tracking_output"
output_video = f"{output_dir}/tracked_video.mp4"

# Create output directory
os.makedirs(output_dir, exist_ok=True)

print(f"Model loaded from: {model_path}")
print(f"Input video: {input_video}")
print(f"Output will be saved to: {output_video}")
```

```python
def process_video_with_tracking():
    """
    Process video with YOLOv11 tracking and save output
    """
    # Run tracking with BoT-SORT
    results = model.track(
        source=input_video,
        tracker="botsort.yaml",  # Use BoT-SORT with ReID
        persist=True,            # Maintain IDs across frames
        conf=0.25,              # Confidence threshold
        vid_stride=1,           # Process every frame
        save=True,              # Save annotated video
        project=output_dir,     # Output directory
        name="run",             # Run name
        show=False,             # Don't show live display
        verbose=True            # Show progress
    )
    
    return results

# Process the video
print("Starting soccer player tracking...")
tracking_results = process_video_with_tracking()
print("Tracking completed!")
```
